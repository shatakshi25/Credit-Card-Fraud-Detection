## **Credit Card Fraud Detection**

The Credit Card Fraud Detection Problem includes modeling past credit card transactions with the knowledge of the ones that turned out to be fraud. This model is then used to identify whether a new transaction is fraudulent or not.

Steps:

*   Import all the required libraries
*   Collection of data

*  Classification of the data into Normal and fraud data

*   Visualization of data - helps to get to know the data and the corresponding features

*   Segregation of the data for training and testing

*   Testing the data and find the best algorithm for the detection


*   Predict the Result

**Segregation of the data for training and testing**

```
from sklearn.model_selection import train_test_split
```

```
#Dividing the data into x and y
x = data.iloc[:,:-1]
y = data['Class']
```

```
#Splitting of data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.35)
```

Import all the libraries required for using different algorithm and for finding out the accuracy score

```
#Defining the libraries required for the algorithms
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```
